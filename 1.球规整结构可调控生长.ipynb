```python
import random
#球的半径
r=6
#基体的尺寸
X=100
Y=100
Z=100
#p1 能控制球的总体积分数
p1=0.5
#把基质填满介质1
dic={}
for i in range(1,X+1):
    for j in range(1,Y+1):
        for k in range(1,Z+1):
            dic[str(i)+','+str(j)+','+str(k)]=str(1)
#先找到核
n1=0
#dic1为球心坐标的字典
dic1={}
lsx=[]
lsy=[]
lsz=[]
for i in range(r+2,X-r):
    for j in range(r+2,Y-r):
        for k in range(r+2,Z-r):
            random_num1 = random.random()
            if random_num1<=p1:
                dic1[str(i)+','+str(j)+','+str(k)]=str(2)
                lsx.append(i)
                lsy.append(j)
                lsz.append(k)
                n1=n1+1
#输出初始核的个数
print('初始结构数量：%s'%n1)
#每一个核进行生长，去除接触的核
#根据判定条件生成球
n2=0
while n2<n1:
    m=0
    break_out_flag1= False
    break_out_flag2= False
    x1=int(lsx[n2])
    y1=int(lsy[n2])
    z1=int(lsz[n2])
    x0=x1-r
    y0=y1-r
    z0=z1-r
    for a in range(2*r+1):
        for b in range(2*r+1):
            for c in range(2*r+1):
                x2=x0+a
                y2=y0+b
                z2=z0+c
                if dic[str(x2)+','+str(y2)+','+str(z2)]==str(1):
                    if ((x2-x1)**2+(y2-y1)**2+(z2-z1)**2)**(1/2)<=r:
                        dic[str(x2)+','+str(y2)+','+str(z2)]=str(2)
                    m=m+1
                else:
                    break_out_flag1 = True
                    dic1.pop(str(x1)+','+str(y1)+','+str(z1))
                    x0=x1-r
                    y0=y1-r
                    z0=z1-r
                    m1=m
                    for a in range(2*r+1):
                        for b in range(2*r+1):
                            for c in range(2*r+1):
                                x2=x0+a
                                y2=y0+b
                                z2=z0+c
                                if m1>0:
                                    m1=m1-1
                                    if dic[str(x2)+','+str(y2)+','+str(z2)]==str(2):
                                        dic[str(x2)+','+str(y2)+','+str(z2)]=str(1)
                                else:
                                    break_out_flag2= True
                                    break
                            if break_out_flag2:
                                break
                        if break_out_flag2:
                            break
                    break
            if break_out_flag1:
                break
        if break_out_flag1:
            break
    n2=n2+1              
#设置输出结构数量和体积分数
print('结构数量：%s'%len(dic1))
n=0
dic2=dic.copy()
for x in range(1,X+1):
    for y in range(1,Y+1):
        for z in range(1,Z+1):
            if dic2[str(x)+','+str(y)+','+str(z)]==str(2):
                n=n+1
n=n/X/Y/Z
print('体积分数：%s'%n)
#给文件命名
name1=len(dic1)
name2=round(n,4)
path1='E:\Regulargrowth\Ball\Ball,'
path=path1+'n='+str(name1)+',volume='+str(name2)+',r='+str(r)+'.dat'
f = open(path,'w')
#写入dat文件中
for i in dic:
    v=dic[i]
    v='           '+v
    s=i.split(',')
    x3=s[0]
    if len(x3)==1:
        x3='           '+x3
    elif len(x3)==2:
        x3='          '+x3
    elif len(x3)==3:
        x3='         '+x3
    y3=s[1]
    if len(y3)==1:
        y3='           '+y3
    elif len(y3)==2:
        y3='          '+y3
    elif len(y3)==3:
        y3='         '+y3
    z3=s[2]
    if len(z3)==1:
        z3='           '+z3
    elif len(z3)==2:
        z3='          '+z3
    elif len(z3)==3:
        z3='         '+z3
    h=x3+y3+z3+v
    f.write(h)
    f.write('\n')
f.close() 
```

```python
#两点关联函数，算密集程度
lsx1=[]
lsy1=[]
lsz1=[]
length=0
num1=0
for i in dic1:
    s=i.split(',')
    x1=int(s[0])
    y1=int(s[1])
    z1=int(s[2])
    lsx1.append(x1)
    lsy1.append(y1)
    lsz1.append(z1)
for i in range(len(lsx1)):
    x1=int(lsx[i])
    y1=int(lsy[i])
    z1=int(lsz[i])
    for j in range(i+1,len(lsx1)):
        x2=int(lsx[j])
        y2=int(lsy[j])
        z2=int(lsz[j])
        l=((x2-x1)**2+(y2-y1)**2+(z2-z1)**2)**(1/2)
        num1=num1+1
        length=length+l
num=length/num1
print('两点关联程度：%s'%num)
```
